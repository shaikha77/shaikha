```python
#Sample code:
import pandas
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import statsmodels.api as statsmodels
df = pandas.read_csv(r'\Users\fatma\Desktop\assigment1\fatma.csv') #enter the path and name of your csv file

#display the correlation coefficient 
corr=stats.stats.pearsonr(df["poverty"],df["smoke"])[0]
print("correlation coefficient=", corr)
#display the scatterplot 
plt.scatter(df["poverty"],df["smoke"])
plt.title ("smoke vs poverty")
plt.xlabel("poverty")
plt.ylabel("smoke")
plt.show()
```
